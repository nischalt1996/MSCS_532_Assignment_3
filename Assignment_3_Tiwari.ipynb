```python
import random

# ============================================================
# PART 1 - IMPLEMENTATION
# Randomized Quicksort
#
# In this implementation, I am selecting the pivot uniformly
# at random from the current subarray. This helps avoid the
# bad behavior that can happen in regular quicksort when the
# pivot choice is poor, especially for already sorted input.
#
# I also want this implementation to behave properly for
# different edge cases such as:
# - empty arrays
# - arrays with one element
# - arrays with repeated values
# - already sorted arrays
# - reverse sorted arrays
# ============================================================


def randomized_quicksort(arr):
    """
    This is the main function the user calls.

    I made this function simple on purpose so that in the notebook
    it is easier to use and easier to explain. Instead of forcing
    the user to pass low and high indexes manually, this function
    handles that setup internally.

    Parameters:
        arr (list): The list of values to be sorted

    Returns:
        list: A new sorted copy of the original list

    Important note:
    I am returning a sorted COPY instead of modifying the original
    list directly. I personally think this is cleaner for testing
    in Jupyter because the original input can still be printed and
    compared with the output if needed.
    """

    # First, create a copy of the input list.
    # This avoids changing the user's original data.
    arr_copy = arr.copy()

    # If the list is empty or has only one element,
    # it is already sorted, so I can return it immediately.
    if len(arr_copy) <= 1:
        return arr_copy

    # Call the recursive helper function on the full array.
    _randomized_quicksort_helper(arr_copy, 0, len(arr_copy) - 1)

    # Return the sorted copy after quicksort finishes.
    return arr_copy


def _randomized_quicksort_helper(arr, low, high):
    """
    This helper function performs the actual recursive quicksort.

    Parameters:
        arr (list): The list being sorted in place
        low (int): Starting index of the current subarray
        high (int): Ending index of the current subarray

    The recursion keeps dividing the array into smaller parts
    around a pivot until each part becomes trivially sorted.
    """

    # Base case:
    # If low is not less than high, then the subarray has
    # either 0 elements or 1 element, so no sorting is needed.
    if low < high:

        # Partition the array around a randomly chosen pivot.
        # After partitioning, the pivot will be in its final
        # correct sorted position.
        pivot_index = _randomized_partition(arr, low, high)

        # Recursively sort the left side of the pivot.
        _randomized_quicksort_helper(arr, low, pivot_index - 1)

        # Recursively sort the right side of the pivot.
        _randomized_quicksort_helper(arr, pivot_index + 1, high)


def _randomized_partition(arr, low, high):
    """
    This function chooses a pivot uniformly at random from the
    subarray arr[low...high], swaps it into the end position,
    and then performs the normal partition process.

    Parameters:
        arr (list): The list being sorted
        low (int): Starting index of current subarray
        high (int): Ending index of current subarray

    Returns:
        int: The final index position of the pivot
    """

    # Choose a random index between low and high, inclusive.
    # This is the key step that makes the quicksort randomized.
    random_pivot_index = random.randint(low, high)

    # Swap the randomly selected pivot with the last element.
    # I am doing this so I can reuse the standard partition logic
    # where the pivot is assumed to be at arr[high].
    arr[random_pivot_index], arr[high] = arr[high], arr[random_pivot_index]

    # Now call the regular partition function.
    return _partition(arr, low, high)


def _partition(arr, low, high):
    """
    This is the standard Lomuto partition scheme.

    The pivot is the element at arr[high].
    After partitioning:
    - all elements less than or equal to the pivot will be on the left
    - all elements greater than the pivot will be on the right

    Parameters:
        arr (list): The list being sorted
        low (int): Starting index of current subarray
        high (int): Ending index of current subarray

    Returns:
        int: The final correct position of the pivot
    """

    # Take the last element as the pivot.
    pivot = arr[high]

    # 'i' will track the boundary of values less than or equal to pivot.
    # At the start, no elements have been placed in the "smaller" region yet.
    i = low - 1

    # Traverse through the subarray from low to high - 1.
    # I stop at high - 1 because arr[high] is the pivot itself.
    for j in range(low, high):

        # If the current element is less than or equal to the pivot,
        # it belongs on the left side of the partition.
        #
        # I am using <= instead of < so repeated values are handled
        # consistently and still get placed into the left region.
        if arr[j] <= pivot:

            # Move the boundary one step to the right because we found
            # one more element that belongs in the smaller/equal region.
            i += 1

            # Swap the current element into the correct side.
            arr[i], arr[j] = arr[j], arr[i]

    # After the loop finishes, place the pivot immediately after the
    # last element that was <= pivot. This puts the pivot in its final
    # sorted position for this partition step.
    arr[i + 1], arr[high] = arr[high], arr[i + 1]

    # Return the pivot's final index so recursion can continue
    # on the left and right subarrays.
    return i + 1


# ============================================================
# SIMPLE TEST CASES
# I am including these because they help show that the code
# works for the exact kinds of cases mentioned in the assignment.
# These outputs are also useful for screenshots later.
# ============================================================

test_cases = {
    "Empty array": [],
    "Single element": [42],
    "Random array": [10, 7, 8, 9, 1, 5],
    "Already sorted": [1, 2, 3, 4, 5, 6],
    "Reverse sorted": [6, 5, 4, 3, 2, 1],
    "Repeated elements": [4, 2, 4, 1, 4, 3, 2, 4]
}

for case_name, case_values in test_cases.items():
    sorted_result = randomized_quicksort(case_values)
    print(f"{case_name}:")
    print(f"Original: {case_values}")
    print(f"Sorted:   {sorted_result}")
    print("-" * 50)
```

This implementation of Randomized Quicksort selects the pivot uniformly at random from the current subarray before partitioning. I used a recursive helper function along with a standard partition procedure. The code also handles important edge cases such as empty arrays, single-element arrays, repeated values, already sorted arrays, and reverse-sorted arrays. I added extensive comments throughout the code so that each step of the logic is easy to follow.

```python
import time
import random

# ============================================================
# Deterministic Quicksort (First element as pivot)
# ============================================================

def deterministic_quicksort(arr):
    arr_copy = arr.copy()
    _deterministic_helper(arr_copy, 0, len(arr_copy) - 1)
    return arr_copy


def _deterministic_helper(arr, low, high):
    if low < high:
        pivot_index = _deterministic_partition(arr, low, high)
        _deterministic_helper(arr, low, pivot_index - 1)
        _deterministic_helper(arr, pivot_index + 1, high)


def _deterministic_partition(arr, low, high):
    # Using FIRST element as pivot (this is the key difference)
    pivot = arr[low]

    left = low + 1
    right = high

    while True:
        # Move left pointer until element > pivot
        while left <= right and arr[left] <= pivot:
            left += 1

        # Move right pointer until element < pivot
        while left <= right and arr[right] >= pivot:
            right -= 1

        # If pointers cross, break
        if left > right:
            break

        # Swap elements
        arr[left], arr[right] = arr[right], arr[left]

    # Place pivot in correct position
    arr[low], arr[right] = arr[right], arr[low]

    return right


# ============================================================
# PERFORMANCE COMPARISON
# ============================================================

def generate_test_cases(n):
    return {
        "Random": [random.randint(0, 1000) for _ in range(n)],
        "Sorted": list(range(n)),
        "Reverse": list(range(n, 0, -1)),
        "Duplicates": [random.choice([1, 2, 3, 4, 5]) for _ in range(n)]
    }


sizes = [100, 500, 1000]

for size in sizes:
    print(f"\nArray Size: {size}")
    test_cases = generate_test_cases(size)

    for case_name, arr in test_cases.items():

        # Measure Randomized Quicksort
        start = time.time()
        randomized_quicksort(arr)
        randomized_time = time.time() - start

        # Measure Deterministic Quicksort
        start = time.time()
        deterministic_quicksort(arr)
        deterministic_time = time.time() - start

        print(f"{case_name}:")
        print(f"  Randomized QS Time:   {randomized_time:.6f} sec")
        print(f"  Deterministic QS Time:{deterministic_time:.6f} sec")
```

```python
# ============================================================
# PART 2 - IMPLEMENTATION
# Hash Table with Chaining (with Dynamic Resizing)
# ============================================================

class HashTable:
    def __init__(self, initial_size=10):
        """
        Initialize hash table with given size.
        Each slot contains a list (chain) to handle collisions.
        """
        self.size = initial_size
        self.count = 0  # number of elements stored

        # Create empty chains
        self.table = [[] for _ in range(self.size)]

    def _hash(self, key):
        """
        Hash function to map key → index
        Using Python's built-in hash and modulo.
        """
        return hash(key) % self.size

    def _load_factor(self):
        """
        Load factor = number of elements / table size
        Used to decide when resizing is needed
        """
        return self.count / self.size

    def _resize(self):
        """
        Double the table size when load factor gets high.
        Rehash all existing elements.
        """
        old_table = self.table
        self.size *= 2
        self.table = [[] for _ in range(self.size)]
        self.count = 0

        # Reinsert all elements into new table
        for chain in old_table:
            for key, value in chain:
                self.insert(key, value)

    def insert(self, key, value):
        """
        Insert key-value pair into hash table.
        Handles collisions using chaining.
        """
        index = self._hash(key)
        chain = self.table[index]

        # Check if key already exists → update
        for pair in chain:
            if pair[0] == key:
                pair[1] = value
                return

        # Otherwise insert new pair
        chain.append([key, value])
        self.count += 1

        # Resize if load factor exceeds threshold
        if self._load_factor() > 0.7:
            self._resize()

    def search(self, key):
        """
        Search for a key and return its value.
        Returns None if not found.
        """
        index = self._hash(key)
        chain = self.table[index]

        for pair in chain:
            if pair[0] == key:
                return pair[1]

        return None

    def delete(self, key):
        """
        Remove a key-value pair from the table.
        """
        index = self._hash(key)
        chain = self.table[index]

        for i, pair in enumerate(chain):
            if pair[0] == key:
                chain.pop(i)
                self.count -= 1
                return

    def display(self):
        """
        Display full table (for debugging and screenshots)
        """
        for i, chain in enumerate(self.table):
            print(f"Index {i}: {chain}")


# ============================================================
# TEST CASES (for screenshots)
# ============================================================

ht = HashTable()

# Insert values
ht.insert("apple", 10)
ht.insert("banana", 20)
ht.insert("orange", 30)
ht.insert("grape", 40)

# Insert duplicate key (should update)
ht.insert("apple", 50)

print("After Insertions:")
ht.display()

print("\nSearch Results:")
print("apple:", ht.search("apple"))
print("banana:", ht.search("banana"))
print("missing:", ht.search("cherry"))

# Delete a key
ht.delete("banana")

print("\nAfter Deletion:")
ht.display()

# Insert many elements to trigger resizing
for i in range(20):
    ht.insert(f"key{i}", i)

print("\nAfter Resizing:")
ht.display()
```

In this implementation, I created a hash table using chaining to handle collisions. Each index in the table stores a list of key-value pairs. I used Python’s built-in hash function combined with modulo to map keys to indices. The implementation supports insert, search, and delete operations. I also included dynamic resizing when the load factor exceeds 0.7, which helps maintain efficiency and reduce collisions.
